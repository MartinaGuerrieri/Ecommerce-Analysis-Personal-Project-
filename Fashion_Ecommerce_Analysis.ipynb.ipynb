# Data Cleaning

```python
#Remove '% off' from column discount
df['Discount'] = df['Discount'].str.replace('% off','')

#Rename column 'Discount' to 'Discount_%'
df.rename(columns = {'Discount':'Discount_%'}, inplace = True)

#Replacing Nan by NaN
df.replace('Nan', np.nan, inplace=True)
```

```python

#Remove the 'Rs\n' from column MRP
df['MRP'] = df['MRP'].str.replace('Rs\n','')
```

```python
#Change MRP datatype to float
df['MRP'] = df['MRP'].astype(float)
```

```python
# Convert the 'Discount_%' column to float
df['Discount_%'] = df['Discount_%'].astype(float)
```

```python
#Changing SellPrice datatype to float
df['SellPrice'] = df['SellPrice'].astype(float)
```

```python
#Splitting Category into ProductCategory and Gender

df[['ProductCategory', 'Gender']] = df['Category'].str.split('-', expand=True)
```

```python
#Creating new colum called Color from Details

df = pd.DataFrame(df)
df['Color'] = df['Details'].str.split().str[-1]
```

```python
df = df.drop('Color', axis=1)
```

```python
# Extracting the first word before the "-" character which will tell me the Product type among the 7 categories

df = pd.DataFrame(df)
df['ProductType'] = df['Details'].str.split(' - ').str[0].str.split().str[-1]
```

```python
#Converting Rupees in Dollars for practice circumnstances (the dataset is from 2022 so I changed it accordingly)

df["MRP_price_in_dollars"] = round(df["MRP"] * 0.012,2)
df["SellPrice_price_in_dollars"] = round(df["SellPrice"] * 0.012,2)
```

#### What are the most common cateories, brandnames and discount? 

```python
# function to generate bar plot 
def count_plots(column, topn=10, y_label=None):
    plt.figure()
    df[column].value_counts().nlargest(topn).plot(kind='bar')
    plt.title(f"{column} Frequency Distribution")
    plt.ylabel(y_label if y_label else 'Count')

# list of attributes and their corresponding y-axis labels
columns = ['ProductCategory', 'BrandName', 'Discount_%']
y_labels = ['Count of Products', 'Count of Brands', 'Count of Discounts']

for column, y_label in zip(columns, y_labels):
    count_plots(column, y_label=y_label)

# To display all the plots together
plt.show()  


# From the ProductCategory Frequency Distribution we can see that the Westernwear and Indianwear significantly prevail over the other categories.
```

```python
# good idea to keep those top 10 brands for future calculations

top_brands = df['BrandName'].value_counts().nlargest(10).index.to_list()
top_brands
```

```python
top_brands_items = df.BrandName[df['BrandName'].isin(top_brands)].count()
top_brands_items
```

```python
# See the proportion of Top Brands Items 

# The 28.4% of our products are from the top brands. 

pie_sizes = []
def pie_charts(part, whole):
    
    pie_sizes = [part, whole-part]
    
    colors = ['Pink', 'blue']
    
    fig, ax = plt.subplots()
    ax.pie(pie_sizes, colors=colors, autopct='%1.1f%%');

pie_charts(top_brands_items, nitems)
plt.title("Proportion of Top Brands Items");
```

```python
# What brands have less than 10 products in database

#Almost 40% of brands have less than 10 products.

rare_brands = df[df.groupby('BrandName')['BrandName'].transform('size') < 10]
nrare_brands = rare_brands['BrandName'].nunique()

pie_charts(nrare_brands, nbrandnames)
plt.title("Propotion of Rare Brands")
```

# Pricing Questions

#### How is the price distribution looks like for categories and top brands?

```python
# Function to generate boxplot
# Since it's an overview we can hide the outliers
def boxplot_plots(column, measure, topn=10):
    plt.figure()
    topn_column_values = df[column].value_counts().nlargest(topn).index.to_list()
    boxplot_df = df[df[column].isin(topn_column_values)]
    ax = sns.boxplot(data=boxplot_df, 
                     x=boxplot_df[column], 
                     y=measure,
                     showfliers = False)
    ax.set(xlabel=None,
           title=f"{column} {measure} distr")
    plt.xticks(rotation = 45)

columns = ['ProductCategory', 'BrandName']

# start with MRP
for column in columns :
    boxplot_plots(column, 'MRP_price_in_dollars')
```

```python
# Continuing with SellPrice_in_dollars
for column in columns:
    boxplot_plots(column, 'SellPrice_price_in_dollars')
```

### How distribution discounts looks like for categories and top brands

```python
for column in columns:
    boxplot_plots(column, 'Discount_%')
```
